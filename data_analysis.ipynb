```python
with open('D:/DataAnalysis/Mod 1 Code & Data Files/albumlist.csv') as csvfile:
    reader=csv.DictReader(csvfile)
        
        
    albums=[] #as we are have created 'albums' from the 'reader' , we don't need to reopen the file for further analysis
    for row in reader:
        
                
                
        albums.append(row)
print('number albums:', len(albums))
```

```python
#earliest release year for any album in dataset
release_years=[ int(row['Year']) for row in albums if row['Year'] ] #for each row/album in our albums list the string 'Year' exist, we can cast it to an int and add it to the new list called release years  
                                                                    # casting means converting from one datatype to anothe. Here string 'Year' is converted into integer                              
    
```

```python
#catching data error 
def is_valid_years(string):
    try:
        year=int(string) #trying to cast the string to integer
    
    except ValueError: # the values that can't be casted to an int
        return False
    else:
        return year >1400 #return the year itself; if it is greater than 1400
    
       
```

```python
release_years=[int(row['Year']) for row in albums if (is_valid_years (row['Year'])) ] #now the list is done for the values of string 'Year' that can be casted to an integer
print (release_years)
```

```python
min_release_date=min(release_years)
print(min_release_date)
```

```python
#sorting list[albums] by artists
sorted_albums= sorted (albums, key=lambda x: x['Artist']) #the optional key 'lambda function' for given parameter x(a row) tells python which column is to be sorted
print(sorted_albums)# the alphabetically sorted artist list 
```

```python
#the most recently released album
valid_albums=[row for row in albums if (is_valid_years (row['Year']))] #it is used to find filter the albums of valid 'Year'
most_recent=max(valid_albums, key= lambda x: x['Year'] ) # the key as lambda function for given parameter 'x' sorts the max value of 'Year' from valid albums
print(most_recent['Album'], most_recent['Artist'], most_recent['Year'])
```

```python
df_cities_1=df[df['city']=='Pittsburgh']
df_cities_1
```

```python
#here, it is determining the individual star ratings of each restaurant
df_cities_1_star=df_cities_1['stars']
df_cities_2_star=df_cities_2['stars']
df_cities_1_star.head(100)
```

```python
#plotting overlapping histograms
plt.hist(df_cities_1_star, #dataframe to plot
         alpha=0.7,    # it will set the opacity of the bars in the histogram
         color='green',
         label='Pittsburgh',
         bins='auto'  #it will set the number of equal  width bins in the range
        )    
plt.hist(df_cities_2_star, #dataframe to plot
         alpha=0.7,    # it will set the opacity of the bars in the histogram
         color='red',
         label='Las Vegas',
         bins='auto'  #it will set the number of equal  width bins in the range
        )    
plt.xlabel('Rating')
plt.ylabel('Number of Rating Scores')
plt.legend(loc='best')
plt.title('Review Distributon of Pittsburgh and Las Vegas')
plt.show()
```

```python
#plotting histograms side by side
plt.hist([df_cities_1_star, df_cities_2_star],
         alpha=0.7,
         color=['red', 'blue'],
         label=['Pittsburgh', 'Las Vegas'],
         bins='auto'
)
plt.xlabel('Rating')
plt.ylabel('Number of Rating Scores')
plt.legend(loc='best')
plt.title('Review Distribution of Pittsburgh and Las Vegas')
plt.show()
```
